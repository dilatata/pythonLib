## 새로운 DataFrame 에 15,18년도 데이터 병합하기


```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('./dataset/aptPriceDataSet/avg15.09.csv', header=2)
```

```python
data['Unnamed: 1'] = np.where(data['Unnamed: 1'].isna(), data['구분'], data['Unnamed: 1'])
data.head()
```

```python
data = data.drop('구분', axis=1)
```

```python
data = data.drop([1])
data = data.drop([9])
data = data.drop([21])
data = data.drop([11])

data.head()
```

```python
data = data.reset_index()
data.drop('index', axis=1, inplace=True)
data.head()
```

```python
month = data.iloc[0,:-3]
month.head()
```

```python
year = data.columns[:-3]
year = pd.Series(year)
print(year).head()
```

```python

for i, y in enumerate(year):
    if i > 1 and i < 14:
        year[i] = ' '.join(['2014년', month[i]])
    elif i >= 14:
        year[i] = ' '.join(['2015년', month[i]])
    elif i == 1:
        year[i] = ' '.join(['2013년', month[i]])
    elif i == 0:
        year[i] = '시구군'

year.head()
```

```python
data = data.iloc[0:,0:-3]
```

```python

data.columns = year
# data.head()
```

```python
data=data.drop([0])
data.head()
```

### 18년도 데이터 수정하기



```python
df = pd.read_csv('dataset/aptPriceDataSet/avg18.07.csv',names=['지역명','규모구분','연도','월','미터분양가격'],encoding='utf-8')
c = df['규모구분'] == '전체'
df2=df[c]
df2['분양가격'] = df2['미터분양가격'].apply(lambda x : int(int(x.replace(',', '')) * 3.3))
del df2['미터분양가격']
del df2['규모구분']
```

```python
df2.head()
df3 = df2.reset_index()
df3.drop('index', axis=1, inplace=True)
df3
```

```python
df3['연월'] = df3['연도']+'년 '+df3['월']+'월'
```

```python
df3
```

```python
df4 = df3.transpose()#행 열 전환
# df.rename(columns=df.iloc[0], inplace=True)
df4.columns = df3['연월']
df4.drop([1])
```
